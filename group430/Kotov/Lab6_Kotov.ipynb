## Метод Ньютона. 
Для получения итерационной формулы этого метода используем следующие эвристические соображения. Запишем для функции $f(x)$ в окрестности точки $x^k$ формулу Тейлора 2-го порядка
$$ f(x)=Q_k(x)+o(||x-x^k||^2) , ||x-x^k||^2 \to 0 $$
$$  Q_k(x)=f(x^k)+(\Delta f(x^k),(x-x^k))+\frac{1}{2}(\Delta^2 f(x^k)(x-x^k),(x-x^k)) $$
В случае, когда гессиан $\Delta^2 f(x^k)$ положительно определен, квадратичная функция $Q_k(x)$ достигает глобального минимума в точке
$$ x^{k+1}=x^k-[\Delta^2 f(x^k)]^{-1}\Delta f(x^k) $$
Отметим, что для квадратичной функции $f(x)$ метод Ньютона сходится за один шаг. Для достаточно гладкой функции $f(x)$ с положительно определенной матрицей Гессе при удачном
выборе начального приближения $x_0$ итерационная последовательность $x_k$ метода Ньютона сходится к точке минимума с квадра-
тичной скоростью. Однако, найти удачное начальное приближение -
задача довольно сложная, требующая определенного искусства. Модифицируя метод Ньютона введением переменного множителя $α_k$,
получают методы спуска, сходящиеся при любом начальном приближении $x$.

### Поиск точки минимума

```python
def func(X):
    return 2*X[0]*X[0]+3*(X[1]-1.5)**2+1
f=func
def niut (f,n,eps):
    X0=np.zeros((n))
    for i in range (n):
        print('Введите X',i)
        X0[i]=input()
    Y0=np.zeros((n,1))
    Y0=X0
    def d2(i,j):
        Y0=X0
        h=1e-6
        u4=f(X0)
        #print(u4)
        #print(Y0)
        Y0[i]=Y0[i]+h
        #print(Y0)
        u3=-f(Y0)
        #print(u3)
        Y0[j]=Y0[j]+h
        #print(Y0)
        u1=f(Y0)
        #print(u1)
        Y0[i]=Y0[i]-h
        #print(Y0)
        u2=-f(Y0)
        #print(u2)
        Y0[j]=Y0[j]-h
        #print(Y0)
        return (u4+u1+u3+u2)/(h*h)
    Y0=X0
    def d(i):
        Y0=X0
        h=1e-6
        v4 =-f(Y0)
        Y0[i]=Y0[i]+h
        v3 =f(Y0)
        return (v4+v3)/h
    from numpy import linalg 
    F=np.zeros((n,n))
    X=np.zeros((n,8))
    P=np.zeros((n,1))
    l=np.zeros((n,1))
    for i in range (n):
        X[i][0]=X0[i]
    for k in range (7):
        for i in range (n):
            l[i]=d(i)
            for j in range (n):
                F[i][j]=d2(i,j)
        if linalg.det(F)==0:
            print('На',k+1,'итерации','матрица F стала вырожденной')
            break
        O=F
        v=linalg.inv(F)
        c=np.dot(v,l)
        for i in range (n):
                X[i][k+1]=X[i][k]-c[i]
        for t in range (n):
            P[t]=X[t][k]
            X0[t]=X[t][k+1]
        if abs(f(X0)-f(P))/(f(X0))<eps and k!=0:
            print('Точка минимума',X0)
            print('Кол-во итераций:',k+1)
            break
    for i in range (n):
        print('Первая производная в точке минимума по X',i,d(i))    
        print('Вторая производная в точке минимума по X',i,d2(i,i))
    X1=np.zeros((n,k+2))
    for i in range(n):
        for j in range (k+2):
            X1[i][j]=X[i][j]
    for i in range (n):
            for j in range (n):
                O[i][j]=d2(i,j)                       
    d=linalg.eigvals(F)
    print('Собственные числа F',d)
    m=max(d)/min(d)
    print('Показатель овражности:',m)
    print(X1)
    return X1
X1=niut(f,2,0.01)
def x(y,i):
    return np.sqrt(0.5*(-3*(y-1.5)**2-1+i))
for i in range (10):
    y=np.linspace(-10,10,10000)
    plt.grid(True)
    plt.plot(y,x(y,i),'green')
    plt.plot(y,-x(y,i),'green')
plt.plot(X1[1],X1[0])
plt.title('Лини уровня и траектория поиска')
plt.show()
```
